Реализуйте функцию вычисления выборки оценок $\theta^*_1, \ldots \theta^*_B$ для оценки $\widehat \theta$ методом бутстрепа. Функция должна принимать на вход выборку, количество бутстрепных выборок $B$ и может принимать дополнительные параметры для вашего удобства. 

```
def bootstrap(sample, B = 6, *args, **kwargs):  
  '''Возвращает выборку оценок размера B. Параметры могут быть разными, необходимые указаны, лишние проглотит'''
  n = sample.shape[0]
  sample_of_thetas = np.array([])
  for b in range(B):
      sum_bsample = 0
      for i in range(n):
          ind = sps.randint(low=0, high = n).rvs()
          sum_bsample += sample[ind]
      sample_of_thetas = np.append(sample_of_thetas, n / sum_bsample) #оцениваем обратное среднее
  return sample_of_thetas
```

Реализуйте вышеупомянутые доверительные интервалы в виде функций, принимающих на вход выборку и  выборку оценок, полученных с помощью бутстрепа. Функции могут принимать дополнительные параметры для вашего удобства. Реализовывать вычисление бутстрепных интервалов для подвыборок размера от 1 до n не нужно, только для всей выборки. Обычный и асимптотический доверительный интервалы лучше также реализовать для всех подвыборок размера от 1 до n, как в предыдущем задании, так как это понадобится дальше.


В данной задаче полагайте $\alpha = 0.9$ (для наглядности).

Сгенерируйте реализации выборки из  $Exp(5)$ размеров 5 и  100 и выведите реализации вышеупомянутых доверительных интервалов для этих двух реализаций. Сделайте **выводы**.

*Обратите внимание на то, как задаются параметры экспоненциального распределеня в `scipy`.*

```
def tochnyi(X, alfa = 0.9):
  n = X.shape[0]
  niz = sps.gamma(a = n, scale = 1).ppf((1-alfa)/2) / np.sum(X)
  verh = sps.gamma(a = n, scale = 1).ppf((1+alfa)/2) / np.sum(X)
  return (round(niz,2), round(verh,2))


def asimpt(X, alfa = 0.9):
  n = X.shape[0]
  means = np.mean(X)
  niz = 1/means - sps.norm().ppf((1+alfa)/2) / ((n**0.5) * means)
  verh = 1/means + sps.norm().ppf((1+alfa)/2) / ((n**0.5) * means)
  return (round(niz,2), round(verh,2))
```

```
def norm_boot(X, boots = None, B = 6, alfa = 0.9):
  if boots is None:
    boots = bootstrap(X, B)
  B = boots.shape[0]
  v_boot = np.var(boots)
  niz = 1/np.mean(X) - sps.norm().ppf((1+alfa)/2) * v_boot**0.5
  verh = 1/np.mean(X) + sps.norm().ppf((1+alfa)/2) * v_boot**0.5
  return (round(niz,2), round(verh,2))

def centr_boot(X, boots = None, B=6, alfa = 0.9):
  if boots is None:
    boots = bootstrap(X, B)
  B = boots.shape[0]
  theta_boot = sorted(boots)
  niz = 2/np.mean(X) - theta_boot[math.ceil(B * (1+alfa)/2) - 1] #(-1) так как индексация массивов идёт с 0
  verh = 2/np.mean(X) - theta_boot[math.floor(B * (1-alfa)/2) - 1]
  return (round(niz,2), round(verh,2))

def kvant_boot(X, boots = None, B = 6, alfa = 0.9):
  if boots is None:
    boots = bootstrap(X, B)
  B = boots.shape[0]
  theta_boot = sorted(boots)
  niz = theta_boot[math.floor(B * (1-alfa)/2) - 1]
  verh = theta_boot[math.ceil(B * (1+alfa)/2) - 1]
  return (round(niz,2), round(verh,2))
```

```
samples = [sps.expon(scale = 1/5).rvs(size = 5), sps.expon(scale = 1/5).rvs(size=100)]

for x in samples:
  print(f'Асимпт. дов. интервал при n = {x.shape[0]}: {asimpt(x)}')
  B = 10
  boots = bootstrap(x, B)
  print(f'Точный дов. интервал при n ={x.shape[0]}: {tochnyi(x)}')
  print(f'Норм. бутстреп. интервал при n = {x.shape[0]}, B = {B}: {norm_boot(x, boots)}')
  print(f'Центр. бутстреп. интервал при n = {x.shape[0]}, B = {B}: {centr_boot(x, boots)}')
  print(f'Квант. бутстреп. интервал при n = {x.shape[0]}, B = {B}: {kvant_boot(x, boots)}')
  print()
```

**Обычные интервалы имеют более менее адекватные границы уже при небольших размерах выборки: они широкие, но охватывают оцениваемое значение. В случае бутстрепных интервалов, при малых количествах бутстрепных выборок все выглядит плохо: интервалы хоть и уже, но иногда не охватывают оцениваемое значение, залазят в отрицательную область, в которой оцениваемый параметр в принципе значений не принимает. Ситуация улучшается при большем количестве бустрепных выборок:**

```
samples = [sps.expon(scale = 1/5).rvs(size = 5), sps.expon(scale = 1/5).rvs(size=100)]
B = 300

for x in samples:
  boots = bootstrap(x, B)
  print(f'Норм. бутстреп. интервал при n = {x.shape[0]}, B = {B}: {norm_boot(x, boots)}')
  print(f'Центр. бутстреп. интервал при n = {x.shape[0]}, B = {B}: {centr_boot(x, boots)}')
  print(f'Квант. бутстреп. интервал при n = {x.shape[0]}, B = {B}: {kvant_boot(x, boots)}')
  print()
```

Сформулируем задачу на математическом языке. 

Пусть $X_1, ..., X_n \sim Bern(p)$ &mdash; проведенные наблюдения, где $X_i = 1$, если в $i$-м испытании случилась неудача после того, как черный кот перебежал дорогу, а $p$ &mdash; неизвестная вероятность такого события. 
Вы хотите проверить гипотезу $\mathsf{H}_0\colon p=1/2$ против альтернативы $\mathsf{H}_1\colon p>1/2$. Нулевая гипотеза означает отсутствие связи между черным котом и неудачей, первая &mdash; что неудача происходит чаще если черный кот перебегает дорогу.

Известно, что $S = \left\{T(x) \geqslant c_\alpha\right\}$, где $T(X) = \sum\limits_{i=1}^n X_i$, является равномерно наиболее мощным критерием для проверки этих гипотез. Это получается аналогично [примеру](https://wiki.yandex.ru/homepage/phystechdatascience-chast-1/vesna-2021/.files/examplernmk-1.pdf), разобранному на wiki. 

Чему при этом равно $c_\alpha$? Как определить p-value?

Оформите это в виде таблицы, можно через `pandas.DataFrame`.

*Замечание:* Пользуйтесь функциями из `scipy.stats`. Внимательно проверьте правильность строгих и нестрогих знаков.

```
n_s  = np.arange(1,35,5)
alfa = 0.05
c_alfa = np.array([])
p_val = np.array([])
for len in n_s:
  if len % 2 == 0: #генерирую выборку из H_0
    p_p = 0.5
  if len % 2 != 0: #из Н1
    p_p = 0.7
  samples = sps.bernoulli(p = p_p).rvs(size = len)
  c_alfa = np.append(c_alfa, round(sps.binom(n = len, p = 0.5).ppf(1 - alfa),2))
  p_val = np.append(p_val, round(1 - sps.binom(n = len, p = 0.5).cdf(np.sum(samples)),2))
criter = p_val < alfa
res = pd.DataFrame({'n':n_s , 'c_alfa': c_alfa, 'p-val': p_val, 'criter': criter})
res
```
